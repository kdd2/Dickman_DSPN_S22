# Day and Person-level Associations Between Social Relationship Characteristics and Sleep in Daily Life

## Background, Aims, and Hypotheses

### Background

Good sleep is a known resillience factor for a variety of mental and physical health conditions including depression (Alvaro et al., 2013), adiposity and metabolic syndrome (Jennings et al., 2007), and heart disease (Wang et al., 2016). There is growing evidence that positive social relationships are linked with improved sleep across the adult lifespan (Gordon et al., 2021; Kent de Grey et al., 2018), representing one pathway through which positive sleep may be promoted. However, thus far, the majority of this work has focused on between-person diferences in sleep characteristics predicted global measures of social relationship characteristics (Gordon et al., 2021). Far less work has examined how day-to-day fluctuations in sleep and social experiences may impact one another. It is possible that such relationships are bidirectional: that is, social relationships during the day impact nightly sleep and nightly sleep experiences impact following day social relationships. In one recent analysis of this sort, Harris and collegues demonstrated evidence that nightly sleep quality predicted next-day affect, belonging, and well-being, while daily wellbeing only predicted nightly sleep quality (Harris et al., 2022). It thus appears likely that bidirectional relationships manifest and that relationships may differ for various types of social relationship characteristics. The present study expands on previous work to investigate how multiple dimentions of daily life social relationships (loneliness, emotional support) relate with multiple dimentions of sleep (sleep quality, sleep duration) between- and within- people. 

In addition, little work thus far has examined whether relationships between social relationships and sleep differ between people depending on personal or circumstantial characteristics. It is possible that bidirectional links between daily social relationship characteristics and sleep differ depending on whether an individual possesses chronic sleep concerns. An individual with insomnia, for example, may be especially socially withdrawn following a night of relatively bad sleep. 

References:
* Alvaro, P. K., Roberts, R. M., & Harris, J. K. (2013). A systematic review assessing bidirectionality between sleep disturbances, anxiety, and depression. Sleep, 36(7), 1059-1068.
* Gordon, A. M., Carrillo, B., & Barnes, C. M. (2021). Sleep and social relationships in healthy populations: a systematic review. Sleep medicine reviews, 57, 101428.
* Harris, P. E., Gordon, A. M., Dover, T. L., Small, P. A., Collins, N. L., & Major, B. (2022). Sleep, Emotions, and Sense of Belonging: A Daily Experience Study. Affective Science, 1-12.
* Jennings, J. R., Muldoon, M. F., Hall, M., Buysse, D. J., & Manuck, S. B. (2007). Self-reported sleep quality is associated with the metabolic syndrome. Sleep, 30(2), 219-223.
* Kent de Grey, R. G., Uchino, B. N., Trettevik, R., Cronan, S., & Hogan, J. N. (2018). Social support and sleep: A meta-analysis. Health Psychology, 37(8), 787.
* Wang, D., Li, W., Cui, X., Meng, Y., Zhou, M., Xiao, L., ... & Chen, W. (2016). Sleep duration and risk of coronary heart disease: a systematic review and meta-analysis of prospective cohort studies. International Journal of Cardiology, 219, 231-239.


### Aims

The present study thus aims to:

**Aim 1:** Test whether average (person-level) social relationship characteristics (loneliness, emotional support) are independantly related to **(1A)** average sleep quality, and **(1B)** average sleep duration.

**Aim 2:** Examine whether daily social relationship characteristics are related with following nights sleep **(2A)** quality, and **(2B)** duration.

**Aim 3:** Examine whether nightly sleep characteristics are related with following day reports of **(3A)** loneliness and **(3B)** emotional support.

**Aim 4:** Test whether daily bidirectional relationships between sleep and social relationship characteristics vary by chronic sleep disurbance status.  

## Variables

**Subject ID** = Subject_ID

**Demographic Characteristics & Covariates**
* Age = Demo_Age; As reported by participants (in years)
* Sex = Sex_bi; As reported by participants (0 = male, 1 = female)
* Race = race_bi; As reported by participants (0 = BIPOC, 1 = white; Due to small sample sizes, if individuals endorsed any racial group other than white or in addition to white they were classified as BIPOC)
* Education = educat; As reported by participants (4 ordinal levels where 1 = High school education or less and 4 is graduate degree)

**Social Relationship Variables**
* Loneliness = loneliness_day; Sum score of two self-report loneliness items measured at hourly intervals on a 1-10 Likert scale. 
> * Loneliness_bp = person-level averages for loneliness across all EMA days. Used in analyses when loneliness is a predictor
> * Loneliness_wp = within-person loneliness calculated as day-level fluctuation in loneliness values around a person's mean. Used in analyses when loneliness is a predictor
* Emotional Support = emosupport_day; Sum score of three self-report emotional support items measured at hourly intervals on a 1-10 Likert scale.
> * Emosupport_bp = person-level averages for emotional support across all EMA days. Used in analyses when emotional support is a predictor
> * Emosupport_wp = within-person emotional support calculated as day-level fluctuation in emotional support values around a person's mean. Used in analyses when emotional support is a predictor

**Sleep**
* Sleep Quality = bodqual; Sleep quality as reported daily by participants upon awakening (1-10 Likert Scale)
> * bodqual_bp = person-level averages for sleep quality across all EMA days. Used in analyses when sleep quality is a predictor
> * bodqual_wp = within-person sleep quality calculated as day-level fluctuation in sleep quality values around a person's mean. Used in analyses when sleep quality is a predictor
* Sleep Duration = Scored_Total_Sleep_Time; Daily total sleep duration as derived through actigraphy data
> * TST_bp = person-level averages for sleep duration across all EMA days. Used in analyses when sleep duration is a predictor
> * TST_wp = within-person sleep duration calculated as day-level fluctuation in sleep duration values around a person's mean. Used in analyses when sleep quality is a predictor

**Chronic Sleep Disturbance** = psqi_disorder; Dichotomous measure of chronic sleep issues endorsed when score of PSQI greater than 5

## Data Organization

### Data Cleaning

Data for this project was pulled from the Study of Health and Interactions in the Natural Environment (SHINE). 391 participants completed 4 laberatory visits over the course of several months. Between visit 3 and 4, participants completed a 10-day long ecological monitoring protocol. As a part of this, participants worse an actigraphy device on the wrist and responded to daily morning and evening surveys for the full 10 day period. For 4 selected days within this 10-day period, participants also completed hourly ecological momentary assessments (EMA). 

The original data was located in several different datasets including:
* An dataset with hourly survey responses collected through Ecological Momentary Assessment over the course of 4 days
* A sleep dataset with daily sleep values (including duration) calculated from actigraphy data
* A demographics dataset with person-level information on demographics characteristics
* A depression dataset with person-level responses on the CES-D
The original datasets had raw variables, extra variables not needed for this project, and were on a variety of different timescales (hourly, day, person-level)

For the MLMs (Aims 2, 3, 4) data needed to be in long format, where each row represented a unique day. Each participant had multiple days of data. For the regression (Aim 1) data needed to be in wide format, where each row represented a unique person. 

The following steps were followed to organize the data for this project:
1. Create new csv files from old datasets for EMA (hourly), actigraphy sleep (daily), demographics (person), and CESD (person)
2. Only keep data that is relevant to my questions in each dataset:
> * **EMA** = Subject_ID, Loneliness_item_1, Loneliness_item_2, Emotional_Support_item_1, Emotional_Support_item_2, Emotional_Support_item_3, Current_Social_Interaction, Date, EMA_monitorday, Type_of_Measurement, Beginning_of_day_sleep_quality
> * **Sleep** = Subject_ID, Date, Duration, Efficiency
> * **Demographics** = Subject_ID, Age, Sex, Education
> * **CESD** = Subject_ID, CESD_Total_Score
3. Create new csv file with day-level aggregates for each of the social relationship characteristic variables and sleep quality
> * 3a) Create hourly mean scores for loneliness (average of Loneliness_item_1 and Loneliness_item_2) & emotional support (average of Emotional_Support_item_1, Emotional_Support_item_2, and Emotional_Support_item_3). Emotional support measures only calculated if the person was in a recent (under 10 minutes) social interaction
> * 3b) Use aggregate() function to create day-level mean values for emotional support and loneliness
> * 3c) Use filter(Type_of_Measurement == Beginning_of_day) function to create new day-level dataset with only subject_ID, Date, and Beginning_of_day_sleep_quality variables
4. Merge new day-level EMA datasets with day-level actigraphy sleep datasets to create a new day-level dataset with all variables included. These datasets are merged by "subject_ID" and "Date"
5. Since not every day in the 10-day sleep period had EMA measurements, we need to **create two new datasets where every sleep measurement is matched with an EMA measurement** (filter out the days that have no EMA matching sleep). The goal was to be prepared to run bi-directional analyses where (1) sleep the previous night predicts social relationships the following day, and (2) where social relationships during a given day predict sleep the following night. To do this, filter() was used in combination with the EMA_monitorday and Date to create two datasets, where:
> * Sleep from previous night match on a line with social relationship characteristics the following day (shine_sleep_to_social)
> * Social relationships characteristics from  day match on a line with sleep from that following night (shine_social_to_sleep)
6. In order to seperate out within- and between- person variance for the MLM models, do the following for each of the two above datasets: 
> * 6a) Create person mean values across all days for each set of predictor variables using aggregate() function
> * 6b) Subtract raw day-level variable values by the person-means to create within-person variables demonstrating variance around person means
> * 6c) Merge between and within person predictor variables with original dataset
7. Create third person-level dataset to complete Aim 1. Person-level mean values (generated using aggregate() function as described above) for all variables combined to create dataset where each row represents a unique participant. 29 individuals were removed for missing actigraphy data. 

```r
library(tidyverse)

sleep_to_social_data <- read.csv("shine_sleep_to_social_datascience.csv")
head(sleep_to_social_data)

social_to_sleep_data <- read.csv("shine_social_to_sleep_datascience.csv")
head(social_to_sleep_data)

aim1.data <- read.csv("shine_personlevel.csv")
head(aim1.data)
```

### Aim 1 Summary & Conclusions:

Aim 1 investigated whether average loneliness and emotional support during daily life are related with average sleep quality and sleep duration in daily life. This aim was tested using linear regression models. Analyses found evidence for independant relationships between loneliness and emotional support ratings and sleep quality ($\beta$ = -0.31, 95% CI: [-0.45, -0.16], $p$ < .001; $\beta$ = 0.20, 95% CI: [0.11, 0.29], $p$ < .001, respectively). These relationships replicate prior literature showing global associations between social relationship characteristics and sleep quality. Analyses found no evidence for independant relationships between loneliness and emotional support ratings and actigraphy-derived sleep duration ($\beta$ = 2.55, 95% CI: [-1.89, 6.98], $p$ = .26; $\beta$ = -0.31, 95% CI: [-3.17, 2.56], $p$ = 83, respectively).

### Aim 2A: Daily relationships between social relationship characteristics and following night sleep quality

```r
library(nlme)

model.2A <- lme(bodqual ~ loneliness_wp + emosupport_wp + Age_c + race_bi + Sex_bi + educat_c + loneliness_bp + emosupport_bp, 
                  data = social_to_sleep_data, 
                  random = ~ 1 + loneliness_wp + emosupport_wp |subjectID, 
                  method="ML", 
                  na.action="na.omit")
summary(model.2A) 
VarCorr(model.2A)
intervals(model.2A)
```

### Aim 2B: Daily relationships between social relationship characteristics and following night sleep duration

```r
model.2B <- lme(Scored_Total_Sleep_Time ~ loneliness_wp + emosupport_wp + Age_c + race_bi + Sex_bi + educat_c + loneliness_bp + emosupport_bp, 
                  data = social_to_sleep_data, 
                  random = ~ 1 + loneliness_wp + emosupport_wp |subjectID, 
                  method="ML", 
                  na.action="na.omit")
summary(model.2B) 
VarCorr(model.2B)
intervals(model.2B, which = "fixed")
```

### Aim 2 Summary & Conclusions:

Aim 2 investigated associations between daily fluctuations in loneliness and emotional support and following night sleep characteristics. This aim was tested using multilevel models. Analyses found no evidence for independant relationships between daily fluctuations in loneliness and emotional support ratings and sleep quality ($\beta$ = -0.00, 95% CI: [-0.34, 0.33], $p$ = .98; $\beta$ = 0.17, 95% CI: [-0.01, 0.36], $p$ = .07, respectively). The between person relationships between average loneliness and amotional support and nightly sleep quality observed in the first set of models are maintained. Analyses likewise found no evidence for independant relationships between daily fluctuations in loneliness and emotional support ratings and actigraphy-derived sleep duration ($\beta$ = -1.86, 95% CI: [-13.27, 9.55], $p$ = .98; $\beta$ = 2.26, 95% CI: [-2.64, 7.15], $p$ = .37, respectively).

Interestingly, model 2A does demonstrate significant random effects of loneliness_wp and emosupport_wp on sleep quality (the confidence intervals do not cross 0), meaning that people vary in the relationship between daily fluctuations in social relationship characteristics and sleep quality. These random effects will be important in regards to aim 4 (further below).

Analyses found no evidence for independant relationships between nightly fluctuations in sleep quality and duration and next-day ratings of loneliness ($\gamma$ = -0.01, 95% CI: [-0.02, 0.01], $p$ = .34; $\gamma$ = -0.00, 95% CI: [-0.00, 0.00], $p$ = .41, respectively). When the impacts of nightly sleep quality and duration were investigated on next-day emotional support, a significant effect of sleep quality emerged ($\gamma$ = -0.04, 95% CI: [-0.07, -0.01], $p$ = .02). Nights with relatively lower sleep quality are associated with relatively greater emotional support. It seems like, on a day to day level, emotional support may play a compensatory effect for poor sleep quality. There was no significant impact of nightly sleep duration on next-day emotional support ($\gamma$ = 0.00, 95% CI: [-0.00, 0.00], $p$ = .11).

### Aim 4 Summary & Conclusions:

 Aim 4 investigated whether the presence of chronic sleep disturbances (defined by a high PSQI score) impacts the bidirectional relationships between between daily fluctuations in sleep and interpersonal relationship chracteristics. This aim was tested using multilevel models. Analyses found support for this moderation hypothesis in two out of four models. More specifically, the present of chronic sleep disturbance appears to moderate the relationship between emotional support and nightly sleep quality. More specifically, it appears that greater emotional support associates with improved sleep quality the following night ONLY in individuals without chronic sleep problems. Similarly, models suggest that the link between poorer nightly sleep quality and next-day loneliness manifests only for individuals with chronic sleep concerns.
